```python
import pandas as pd
dataset = pd.read_csv('data/titanic.csv')
print(dataset)
```

```python
rows, cols = dataset.shape
print('Jumlah baris', rows)
print('Jumlah kolom', cols)
```

```python
data = pd.DataFrame(dataset, columns=['Name', 'Sex', 'Age', 'Pclass', 'Fare'])
print(data)
```

```python
classs = pd.DataFrame(dataset, columns=['Survived'])
print(classs)
```

```python
data['Relatives'] = dataset['SibSp'] + dataset['Parch']
print(data)
```

```python
highclass = data['Pclass'].value_counts()[1]
print('Highclass', highclass)
midclass = data['Pclass'].value_counts()[2]
print('Midclass', midclass)
lowclass = data['Pclass'].value_counts()[3]
print('Lowclass', lowclass)
```

```python
count = data['Sex'].value_counts()['male']
print('Male', count)
count = data['Sex'].value_counts()['female']
print('Female', count)
```

```python
print('Highclass')
survived = dataset[(dataset['Pclass'] == 1)&(dataset['Survived']==1)]
print('Selamat Highclass', survived['Survived'].count())
notSurvived = dataset[(dataset['Pclass'] == 1)&(dataset['Survived']==0)]
print('Tidak Selamat Highclass', notSurvived['Survived'].count())

print('Midclass')
survived = dataset[(dataset['Pclass'] == 2)&(dataset['Survived']==1)]
print('Selamat Midclass', survived['Survived'].count())
notSurvived = dataset[(dataset['Pclass'] == 2)&(dataset['Survived']==0)]
print('Tidak Selamat Midclass', notSurvived['Survived'].count())

print('Lowclass')
survived = dataset[(dataset['Pclass'] == 3)&(dataset['Survived']==1)]
print('Selamat Lowclass', survived['Survived'].count())
notSurvived = dataset[(dataset['Pclass'] == 3)&(dataset['Survived']==0)]
print('Tidak Selamat Lowclass', notSurvived['Survived'].count())
```
